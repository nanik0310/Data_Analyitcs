```python
mean_units = df["NUMBER_OF_UNITS"].mean()
mean_employment = df["TOTAL_EMPLOYMENT"].mean()

print(f"Mean of Number of Units: {mean_units}")
print(f"Mean of Total Employment: {mean_employment}")
```

```python
median_units = df["NUMBER_OF_UNITS"].median()
median_employment = df["TOTAL_EMPLOYMENT"].median()

print(f"Median of Number of Units: {median_units}")
print(f"Median of Total Employment: {median_employment}")
```

```python
from scipy.stats import ttest_ind

group_high = df[df["TOTAL_EMPLOYMENT"] > 500000]["NUMBER_OF_UNITS"]
group_low = df[df["TOTAL_EMPLOYMENT"] <= 500000]["NUMBER_OF_UNITS"]

t_stat, p_value = ttest_ind(group_high, group_low, equal_var=False)

print(f"T-test statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Significant difference between high and low employment districts.")
else:
    print("No significant difference between groups.")
```

```python
from scipy.stats import chi2_contingency

df["EMPLOYMENT_CATEGORY"] = df["TOTAL_EMPLOYMENT"].apply(lambda x: "High" if x > 500000 else "Low")

contingency_table = pd.crosstab(df["EMPLOYMENT_CATEGORY"], df["NUMBER_OF_UNITS"] > 25000)

chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Employment category and number of units are dependent.")
else:
    print("No significant dependency between employment and number of units.")
```
