```python
data = {'PP': [1, 2, 3, 4, 5, 6, 7], 'SECTION': ['ALPHA', 'BETA', 'GAMMA', 'DELTA', 'SIGMA' , 'ZETA', 'EPSILON']}
df = pd.DataFrame(data)
if not pd.api.types.is_numeric_dtype(df['SECTION']):
    df['SECTION'] = pd.Categorical(df['SECTION']).codes
df.plot.scatter(x='PP', y='SECTION', color='blue', s=30)
plt.title("Scatter Plot for PP")
plt.show()
```

```python
data = {'BEEE': [10, 20, 30, 40, 50, 60, 70], 'SECTION': ['ALPHA', 'BETA', 'GAMMA', 'DELTA', 'SIGMA' , 'ZETA', 'EPSILON']}
df = pd.DataFrame(data)
if not pd.api.types.is_numeric_dtype(df['SECTION']):
    df['SECTION'] = pd.Categorical(df['SECTION']).codes
df.plot.scatter(x='BEEE', y='SECTION', color='Green', s=20)
plt.title("Scatter Plot for BEEE")
plt.show()
```

```python
data = {'FL': [1, 2, 3, 4, 5, 6, 7], 'SECTION': ['ALPHA', 'BETA', 'GAMMA', 'DELTA', 'SIGMA' , 'ZETA', 'EPSILON']}
df = pd.DataFrame(data)
if not pd.api.types.is_numeric_dtype(df['SECTION']):
    df['SECTION'] = pd.Categorical(df['SECTION']).codes
df.plot.scatter(x='FL', y='SECTION', color='cyan', s=30)
plt.title("Scatter Plot for FL")
plt.show()
```
